```python
import requests
from bs4 import BeautifulSoup
print('hello')
url = "https://pubchem.ncbi.nlm.nih.gov/compound/1"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Extract the name of the compound
name = soup.find("h1", class_="pagetitle").text
print("Compound name:", name)

# Extract the molecular formula
molecular_formula = soup.find("div", class_="infoleft").text
print("Molecular formula:", molecular_formula)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://pubchem.ncbi.nlm.nih.gov/compound/1"

driver = webdriver.Chrome('./chromedriver_win32/chromedriver')
driver.get(url)

# Wait for the page to load
wait = WebDriverWait(driver, 50)
element = wait.until(EC.presence_of_element_located((By.ID, "compound-info-right")))

# Extract the boiling point
boiling_point = driver.find_element_by_xpath("//td[text()='Boiling Point']/following-sibling::td").text
print("Boiling point:", boiling_point)

driver.quit()
```

```python
#now gonna figure out how to put everything I want from the first compound, into a data frame, and then go from there. From the 
#pandas documentation, using a dictionary like d = {'col1': [1, 2], 'col2': [3, 4]} makes it easy to create a dataframe.
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

n = 2
elements_repr = ['C']
d = {'CID': range(1,10**n), 'smiles_repr':[], 'selfies_repr':[], 'h_acc':[], 'h_don':[], 'C':[]}
for i in d['CID']:
        
    compound = pcp.Compound.from_cid(i)
    d['smiles_repr'].append(compound.isomeric_smiles)
    d['selfies_repr'].append(sf.encoder(compound.isomeric_smiles))
    d['h_acc'].append(compound.h_bond_donor_count)
    d['h_don'].append(compound.h_bond_acceptor_count)    
    
    #print(compound.molecular_formula)
    
    for j in ( set(compound.elements) - set(elements_repr) ):
        d[j] = [0]*len(d['C'])
        elements_repr.append(j)        
        
    for k in elements_repr:
        d[k].append(compound.elements.count(k))        
        
    #if 'C' in d.keys():
        #print(d['C'])

    #if 'P' in d.keys():
        #print(d['P'])
        
    #print('new compound \n\n')
#for i in d:
#    print(len(d[i]))
#    print (i, d[i])


df = pd.DataFrame(data = d)
df
```
